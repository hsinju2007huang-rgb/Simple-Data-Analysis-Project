```python
import pandas as pd

grades=pd.read_csv('C:\\Users\\USER\\Downloads\\ultimate_student_productivity_dataset_5000.csv')

print(grades.head())
```

```python
grades.describe()
```

```python
grades.info()
```

```python
#filter non-numerical columns as well as complicated columns with data to analyze correlation
columns_to_drop = ['student_id', 'gender', 'academic_level', 'internet_quality', 'gaming_hours', 'part_time_job', 'upcoming_deadline', 'social_media_hours','exercise_minutes']
grades = grades.drop(columns=columns_to_drop)
print(grades.head())
```

```python
grades.corr()
```

```python
# What I plan to observe is how various factors such as study hours, sleep hours, etc affects the grades of students.
# From the above, only focusing on exam_score, comparing with the other columns, I can see that productivity_score has the highest positive correlation with exam_score at 0.886. This indicates that as productivity_score increase, exam_score tends to increase as well. Some other factors to note are focus_index. (They are factors that are above 0.6 correlation with exam_score)
# There are no obvious negative correlations with exam_score, however, screen_time_hours and caffeine_intake have negative correlations (though it is not a strong correlation therefore there is no reason to believe that they are significant factor affecting exam scores.)

#Now I shall visualize these correlations using scatter plots to better understand the relationships between exam_score and the other significant factors.
import matplotlib.pyplot as plt
grades.plot.scatter(x='exam_score', y='productivity_score', title='Productivity Score vs Exam Score')
```
