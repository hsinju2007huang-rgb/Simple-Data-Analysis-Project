```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

grades = pd.read_csv('C:\\Users\\USER\\Downloads\\ultimate_student_productivity_dataset_5000.csv')
print(grades.head())
```

```python
grades.info()
```

```python
Grades = ['exam_score','screen_time_hours','study_hours']
mean=grades[Grades].mean()
std=grades[Grades].std()
print(mean)
print(std)

# To find mean and standard deviation. The describe() function also does the same.
```

```python
exam_scores_mean = grades['exam_score'].mean()
exam_scores_std = grades['exam_score'].std()
data = np.random.normal(exam_scores_mean, exam_scores_std, 5000)

plt.hist(data)
plt.title('Normal Distribution of Exam Scores for 5000 Students')
```

```python
# For less blocky histogram, we can increase the number of bins.
plt.hist(data, bins=1000)
plt.title('Normal Distribution of Exam Scores for 5000 Students')
```

```python
a=grades['online_classes_hours']
b=grades['social_media_hours']
c=grades['gaming_hours']
d=grades['screen_time_hours']
checking = a+b+c==d
print(checking)

#I guess this means that the online classes hours, social media hours and gaming hours does not sum up to screen time hours. There may be other factors contributing to screen time.
```

```python
print(checking.value_counts())

#The previous statement does not apply for 4 other values in the dataset however. Therefore for those 4 students, sum of online classes hours, social media hours and gaming hours is equal to screen time hours. For the rest of the students, there are other factors contributing to screen time hours.
```
